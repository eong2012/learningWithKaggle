```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline 
```

```python
df_train = pd.read_parquet('/home/ubuntu/data/sar_ukho/intermediate_parquets/')
```

```python
df_train.head()
```

```python
df_train.label.value_counts()
```

```python
df_oil = df_train[df_train.label=='oil_and_gas_infrastructure']
df_turbine = df_train[df_train.label=='turbine']
df_other = df_train[df_train.label=='other']
```

```python
len(df_oil), len(df_turbine), len(df_other)
```

```python
plt.figure(figsize=(20,10))
plt.plot(list(range(0, len(df_other))), df_other.distance_to_land, 'o', ms=10, label="other dist to land")
plt.plot(list(range(0, len(df_oil))), df_oil.distance_to_land, 'o', ms=10, label="oil dist to land")
plt.plot(list(range(0, len(df_turbine))), df_turbine.distance_to_land, 'o', ms=10, label="turbine dist to land")
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102),
           ncol=1, mode="expand", borderaxespad=0.)
```

```python
plt.figure(figsize=(20,10))
plt.plot(list(range(0, len(df_other))), df_other.water_depth, 'o', ms=10, label="other water depth")
plt.plot(list(range(0, len(df_oil))), df_oil.water_depth, 'o', ms=10, label="oil water depth")
plt.plot(list(range(0, len(df_turbine))), df_turbine.water_depth, 'o', ms=10, label="turbine water depth")
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102),
           ncol=1, mode="expand", borderaxespad=0.)
```

```python
plt.figure(figsize=(20,10))
plt.plot(list(range(0, len(df_other))), df_other.incidence_angle, 'o', ms=15, label="other incidence_angle")
plt.plot(list(range(0, len(df_oil))), df_oil.incidence_angle, 'o', ms=15, label="oil incidence_angle")
plt.plot(list(range(0, len(df_turbine))), df_turbine.incidence_angle, 'o', ms=15, label="turbine incidence_angle")
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), 
           ncol=1, mode="expand", borderaxespad=0.)
```
